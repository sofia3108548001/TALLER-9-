## DATOS ESPACIALES 
    Un dato espacial es aquel que no puede ser asociado a una posicion en el espacio 
    Ya sea en un plano 2D, 3D o n-dimensional 

```python
import matplotlib.pyplot as plt

#simular un conjunto de datos que represente la elevacion puntual en una terreno 
#crear un conjunnto de datos de prueba 
elevacion= np.random.randint(0, 3000, size=(100,100))
elevacion 
```

```python
plt.imshow(elevacion, cmap= "terrain")
plt.colorbar(label="Altura (m)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Suponiendo que tienes una lista de estaturas en metros llamada "estaturas"
estaturas = ([1.88, 1.60, 1.60, 1.60, 1.60, 1.60, 1.65, 1.59, 1.75, 1.59, 1.72, 1.74, 1.80, 1.75, 1.75, 1.75, 1.69, 1.60])

# Crea un histograma de las estaturas
plt.hist(estaturas, bins=10)

# Agrega etiquetas a los ejes
plt.xlabel("Estatura (metros)")
plt.ylabel("Frecuencia")

# Agrega un título a la gráfica
plt.title("Estatura del grupo")

# Muestra la gráfica
plt.show()
```
